```
from nltk.corpus import names
all_names = set(names.words())

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```
Train_csv = pd.read_csv('/content/drive/My Drive/TRAIN.csv')
Train_csv.head()
```

```
Train_csv.drop('User', axis=1, inplace=True)
Train_csv.drop('Unnamed: 0', axis=1, inplace=True)
Train_csv.drop('Label', axis=1, inplace=True)

Train_csv.head()
```

```
def clean_text(doc):
  cleaned_post = []
  for word in doc.split():
    word = word.lower()
    if word.isalpha() and word not in all_names and len(word) > 2:
      cleaned_post.append(lemmatizer.lemmatize(word))
  return ' '.join(cleaned_post)
```

```
clean_posts = []
for i in range(len(Train_csv.Post)):
  clean_posts.append(clean_text(Train_csv.Post[i]))
x = pd.Series(clean_posts)
Train_csv['Post'] = x
Train_csv
```

```
term_docs_train = []
for i in range(len(Post_1)):
  word_col_dic = {}
  for doc in Post_1[i].split():
    if not word_col_dic.get(doc, False):
      word_col_dic[doc] = 1
    else:
      word_col_dic[doc] += 1
  term_docs_train.append(word_col_dic)
print(term_docs_train)
print(len(term_docs_train))
```

```
def get_label_index(Dataframe):
  label_index = {0:[], 1:[], 2:[], 3:[], 4:[]}
  posts = Dataframe.Post.values
  LabelNumbers = Dataframe.LabelNumber.values
  for idx, label in enumerate(LabelNumbers):
    label_index[label].append(idx)
  return label_index
label_index = get_label_index(Train_csv)
print(label_index)
```

```
def get_prior(label_index):
  prior = {label: len(index) for label, index in label_index.items()}
  total_sum = sum(prior.values())
  for i in prior.keys():
    prior[i] /= total_sum
  return prior
prior = get_prior(label_index)
prior
```

```
# 시간이 좀 걸림
def get_likelihood(term_doc_matrix, label_index, smoothing, word_key, DataFrame):
  likelihood = {}
  for i in range(5):
    likelihood[i] = {}
  number_of_words = [0, 0, 0, 0, 0]
  for label_number in label_index.keys():
    for post_idx in label_index[label_number]:
      number_of_words[label_number] += (sum(term_doc_matrix[post_idx].values()) + smoothing)
  # print(number_of_words) [15366, 31914, 75648, 41961, 41961]
  for word in word_key:
    likelihood[0][word] = 0
    likelihood[1][word] = 0
    likelihood[2][word] = 0
    likelihood[3][word] = 0
    likelihood[4][word] = 0
    for post_idx in range(len(term_doc_matrix)):
      label_no = int(DataFrame.iloc[post_idx].LabelNumber)
      likelihood[label_no][word] += term_doc_matrix[post_idx].get(word, 0) # 해당 포스트의 해당 단어 개수
    for i in range(5):
      likelihood[i][word] += smoothing
      likelihood[i][word] /= number_of_words[i]
  print(likelihood[0])
  return likelihood

likelihood = get_likelihood(term_docs_train, label_index,1, word_key, Train_csv)
```

```
from decimal import Decimal
def get_posterior(term_doc_matrix, likelihood, Post_1):
  posteriors = []
  for post in Post_1:
    prob_of_0 = 0
    prob_of_1 = 0
    prob_of_2 = 0
    prob_of_3 = 0
    prob_of_4 = 0
    y = {}
    for word in post.split():
      try:
        prob_of_0 += Decimal(f'{np.log(likelihood[0][word])}')
        prob_of_1 += Decimal(f'{np.log(likelihood[1][word])}')
        prob_of_2 += Decimal(f'{np.log(likelihood[2][word])}')
        prob_of_3 += Decimal(f'{np.log(likelihood[3][word])}')
        prob_of_4 += Decimal(f'{np.log(likelihood[4][word])}')
      except:
        pass
    prob_of_0 += Decimal(f'{np.log(prior[0])}')
    prob_of_1 += Decimal(f'{np.log(prior[1])}')
    prob_of_2 += Decimal(f'{np.log(prior[2])}')
    prob_of_3 += Decimal(f'{np.log(prior[3])}')
    prob_of_4 += Decimal(f'{np.log(prior[4])}')
    sum_of_prob = Decimal(f'{np.exp(prob_of_0) + np.exp(prob_of_1) + np.exp(prob_of_2) + np.exp(prob_of_3) + np.exp(prob_of_4)}')
    y[0] = float(np.exp(prob_of_0) / sum_of_prob)
    y[1] = float(np.exp(prob_of_1) / sum_of_prob)
    y[2] = float(np.exp(prob_of_2) / sum_of_prob)
    y[3] = float(np.exp(prob_of_3) / sum_of_prob)
    y[4] = float(np.exp(prob_of_4) / sum_of_prob)
    posteriors.append(y)
  return posteriors
posteriors = get_posterior(term_docs_train, likelihood, Post_1)
print(posteriors)
```

```
import numpy as np
```

PCA(Principle Component Analysis)

```
x1 = np.arange(0, 10, .2)
x2 = x1 + np.random.randn(len(x1))
```

```
X = np.c_[(x1, x2)]
good = (x1 > 5) | (x2 > 5) # some arbitrary classes
bad = ~good # to make the example look good
```

```
good, bad
```

```
from sklearn import linear_model, decomposition, datasets
pca = decomposition.PCA(n_components=1)
Xtrans = pca.fit_transform(X)

pca.explained_variance_ratio_
```

LDA(Linear Discriminant Analysis)

```
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

lda_inst = LDA(n_components=1)
Xtrans = lda_inst.fit_transform(X, good)
print(Xtrans)
```

```
# PCA는 비지도, LDA는 지도학습
# 샘플이 적을때는 PCA가 잘 동작
# 샘플이 충분하면 LDA가 잘 동작
```

MDS(Multidimensional Scaling)

```
X = np.c_[np.ones(5), np.full(5, 2), np.full(5, 10)].T
print(X)
```

```
from sklearn import manifold

mds = manifold.MDS(n_components=3)
Xtrans = mds.fit_transform(X)
Xtrans
```

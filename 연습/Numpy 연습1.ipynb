```
import numpy as np
```

```
def np_print(arr):
  text = '''
  type: {}
  shape: {}
  size: {}
  ndim: {}
  dtype: {}
  array data:\n{}'''.format(type(arr), arr.shape, arr.size, arr.ndim, arr.dtype, arr)
  print(text)
```

```
t = [1, 2, 3, 4.5]
t = np.array(t)
np_print(t)
```

```
t = t.astype(np.int64)
np_print(t)
```

```
t = [1, 2, '3', 4.5]
t = np.array(t)
np_print(t)
```

```
t = [1, 2, '3', 4.5]
t = np.array(t, dtype=np.int32)
np_print(t)
```

```
t = [1, 2, '3', 4.5]
t = np.array(t, dtype=np.object)
np_print(t)
```

```
t = [1, 2, 3, 4]
t = np.array(t, dtype=np.float64)
np_print(t)
```

```
t = [1, 2, 3, 4]
t = np.array(t)
np_print(t)
```

```
arr = np.zeros((3,4))
np_print(arr)
```

```
arr = np.ones((3,4))
np_print(arr)
```

```
arr = np.ones((3,4),dtype=int)
np_print(arr)
```

```
arr = np.full((3,4),7)
np_print(arr)
```

```
arr = np.eye(4)
np_print(arr)
```

```
arr = np.eye(4,5)
np_print(arr)
```

```
arr = np.eye(4,3)
np_print(arr)
```

```
arr = np.empty((3,2))
# C처럼 쓰레기값이 들어가 있음, 그러나 이미 저장된 자료를 가져오기만 하는거라 속도가 제일 빠름
np_print(arr)
```

```
arr = np.ones_like(arr)
np_print(arr)
```

```
arr = np.linspace(1, 10, 11) # 이건 잘알지
np_print(arr)
```

```
arr = np.linspace(1,10)
np_print(arr)
```

```
arr = np.linspace(1, 10, 10, endpoint=False) # 이건 몰랐지
np_print(arr)
```

```
arr = np.linspace(1, 10, 10, retstep=True) # tuple을 반환 (배열객체, 간격값)
print(arr)
np_print(arr[0])
```

```
arr = np.arange(5, 0, -1) # 파이썬 range랑 똑같네
np_print(arr)
```

```
arr = np.logspace(0, 1, 10, base=3) # lin이랑 똑같은데 로그스페이스임 endpoint 이런거 다 있음 base로 진수 지정 가능 기본값 10
np_print(arr)
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
mean = 50
std = 3
arr = np.random.normal(mean, std, (10000,)) # 평균과 표준편차를 주고 정규분포함수에서 난수 생성
plt.hist(arr, bins=100) # bins: 몇개의 구간으로 나눠서 각 구간에 난수가 몇개 있는지 센다
plt.show()
```

```
arr = np.random.rand(10000) # rand는 [0,1) 에서 균등분포
plt.hist(arr, bins=100)
plt.show()
```

```
# randn은 표준정규분포에서 난수추출!!
# 제일 많이 사용함!!
arr = np.random.randn(10000)
plt.hist(arr, bins = 100)
plt.show()
```

```
arr = np.random.randint(-100, 100, (10000,)) # randint는 주어진 범위에서 균등추출
plt.hist(arr, bins = 100)
plt.show()
```

```
arr = np.random.random((10000,)) # random은 [0,1)에서 균등추출
plt.hist(arr, bins = 100)
plt.show()
```

```
"""
정리하자면
rand, randint, random은 균등추출
normal, randn은 정규분포 추출

rand, randn, random은 표준값 혹은 0~1값
randint, normal은 범위 지정

rand랑 random 다른점은 rand는 1차원만 되고 random은 다차원도 되는듯
np.random.seed(1) 하면 랜덤값 고정 쌉가능
"""
```

```
x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6], [7, 8]])

v = np.array([9, 10])
w = np.array([11, 12])
```

```
x = np.array([[1, 2], [3, 4]])

print(np.sum(x))
print(np.sum(x, axis = 0)) # 각 col들을 더한 값
print(np.sum(x, axis = 1)) # 각 row들을 더한 값
```

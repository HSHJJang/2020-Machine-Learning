```
%%time
arr = np.arange(12)
print(arr)

arr2 = np.reshape(arr, (4, 3)) # arr.reshape(4,3)도 가능
print(arr2)

print()
arr[1] = 100
print(arr)
print(arr2)

print(arr2.base is arr)
print(arr.base is arr2)
```

```
# call by reference가 아니라 value이고 싶다면? 복사본 이용하자!
arr3 = arr.reshape(4,3).copy()
print(arr3)

print(arr3.base is arr)
```

```
# 데이터가 너무 많아서 몇행 몇열로 설정해야 할지 모르겠다면? -1을 이용 그리고 형태가 맞아야함!!
arr2 = arr.reshape(3,-1)
print(arr2)
```

```
# 다차원 배열을 1차원으로 만들기 --> ravel()
arr4 = np.ravel(arr3)
print(arr4)
```

resize()

```
# reshape와 다르게 형태가 달라도 error가 발생하지 않고 수가 부족하면 배열의 처음부터 다시한다
arr = np.arange(10)

# 야 이거 좀 다른데 공부좀 해야할듯
arr2 = np.resize(arr, (2, 7))
print(arr)
arr.resize(2,7)
print(arr)
print(arr2)
```

array 결합하기

```
# 가로방형으로 결합하기: np.hstack() 열 개수는 노상관
np.random.seed(1111)
arr1 = np.random.randint(0, 10, (2,3))
print(arr1)
arr2 = np.random.randint(0, 10, (2, 5))
print(arr2)
result = np.hstack((arr1, arr2))
result
```

```
# 세로로 결합하기: np.vstack(tuple) 행 개수는 노상관
arr1 = np.random.randint(0, 10, (2,3))
print(arr1)
arr2 = np.random.randint(0, 10, (3, 3))
print(arr2)
result = np.vstack((arr1, arr2))
result
# 둘다 지원하는건 np.concatenate()이다. 기본값은 axis=0 세로길이 늘어나는거고 axis=1해주면 가로가 늘어남
```

Fancy indexing

```
arr = np.arange(12).reshape(3,4)
print(arr)
```

```
print(arr[2,2]) # indexing
print(arr[1:2, 2]) # slicing
print(arr[[0,2], 2]) # 0, 2행의 2열을 도출
print(arr[2, [0,2]])
print(arr[[0,2],1:])
```

배열 연산

```
np.random.seed(1)
arr1 = np.arange(1,7).reshape(2,3)
arr2 = np.random.randint(1, 10, (2,3))
print(arr1+arr2)
```

```
print(arr1 + 10) # broadcasting
```

```
# 그밖에 np.add, np.subtract, np.divide, np.sqrt, np.dot이 있다
arr3 = arr1.reshape(3,2).copy()
np.dot(arr1, arr3) # 2x3 . 3x2
```

```
print(arr1)

np.sum(arr1), arr1.sum()
```

```
np.cumsum(arr1)
```

```
# 여러 통계값들
np.mean(arr1), np.max(arr1), np.min(arr1), np.std(arr1)
```

```
np.sqrt(arr1), np.log10(arr1)
```

```
print(arr1)
arr1.sum(), arr1.sum(axis=0), np.sum(arr1, axis=1) # 0이 행방향, 1이 열방향이라는데 감이 잘 안온다
```

delete

```
# np.delete(): axis를 기준으로 행과 열을 삭제
# 지정하지 않으면 1차배열로 변환된 후 해당요소 삭제
# 원본은 유지, 삭제된 복사본을 return한다
arr = np.arange(12).reshape(3,4)
print(arr,'\n')

print(np.delete(arr,1),'\n')

print(np.delete(arr,1, axis=0),'\n')

print(np.delete(arr,1, axis=1))
```

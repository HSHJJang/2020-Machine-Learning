```
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
plt.figure(figsize=(10,5)) # 표의 크기
plt.title("Plot")
plt.plot([1,4,9,16])
plt.show() # 없어도 됨
```

```
plt.figure(figsize=(10,7))
plt.title("x ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```

```
plt.figure(figsize=(10,7))
plt.title("'rs--' plot")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--') # bs--는 불루 #yo--는 엘로운데 점이 동그라미 -하나면 실선
```

```
plt.figure(figsize=(10,7))
plt.title("x-axis, y-axis limit")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 
         c = "b", lw = 5, ls = "--", marker = "o", ms = 15, mec = "g", mew = 5, mfc = "r")
#      왜 여기 설명 안해줘?
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.show()
```

```
import numpy as np
```

```
plt.figure(figsize=(10,7))
X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
```

```
plt.figure(figsize=(10, 7))
t = np.arange(0., 5., .2)
plt.title("Multiple plots")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()
```

```
plt.figure(figsize=(10, 7))
plt.plot([1, 4, 9, 16],
         c = "b", lw = 5, ls = "--", marker = "o", ms = 15, mec = "g", mew = 5, mfc = "r")
plt.plot([9, 16, 4, 1],
         c = "r", lw = 3, ls = ":", marker = "s", ms = 10, mec = "m", mew = 5, mfc = "c")
plt.show()
```

```
plt.figure(figsize=(10, 7))
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, ls = "--", label = "cosine")
plt.plot(X, S, ls = ":", label = "sine")
plt.legend(loc = 2) # 0,1은 우상단 2는 좌상단 3은 좌하단
```

```
plt.figure(figsize=(10, 7))
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label = "cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
```

```
plt.figure(figsize=(10, 7))
x1 = np.linspace(0.0, 5.)
x2 = np.linspace(0., 2.)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1) # row, col, num
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplot')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```
